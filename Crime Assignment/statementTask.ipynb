```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import folium as fo
from folium.plugins import MarkerCluster
```

```python
data = pd.read_csv("app/Crime_Data_from_2010_to_2019.csv", delimiter=",")
```

```python
crimes = pd.DataFrame(data, columns=['LAT', 'LON', 'DATE OCC', 'VICT Sex', 'Crm Cd Desc', 'Cross Street'])


crimes.rename(columns={'DATE OCC': 'Date'}, inplace=True)
crimes.rename(columns={'VICT Sex': 'Gender'}, inplace=True)
crimes.rename(columns={'Crm Cd Desc': 'Category'}, inplace=True)
crimes.rename(columns={'Cross Street': 'Area'}, inplace=True)

limit = 100
crimes = crimes.iloc[0: limit]

#sexua = (crimes['Gender']== 'F') | (crimes['Gender'] == 'M')

#crimes = crimes[sexua].iloc[0: limit]
```

```python
map = fo.Map()

map
```

```python
lat = 34.0989
lon = -118.3277
```

```python
la_map_default = fo.Map(location=[lat, lon], zoom_start=14)

#la_map = fo.Map(location=[lat, lon], zoom_start=14, tiles='cartodbpositron')

la_map = fo.Map(location=[lat, lon], zoom_start=14, tiles='Stamen Toner')

la_map
```

```python
incidents = fo.map.FeatureGroup()

for lat, lon in zip(crimes['LAT'], crimes['LON']):
    incidents.add_child(
        fo.features.CircleMarker(
            [lat, lon],
            radius = 5,
            color = "yellow",
            filled = True,
            fill_color = 'red',
            fill_opacity = 0.6
        )
    )
    
latitudes = list(crimes['LAT'])
longitudes = list(crimes['LON'])
labels = list(crimes['Category'])

groups = fo.Map(Location=[crimes['LAT'].mean(), crimes['LON'].mean()])

marker_cluster = MarkerCluster().add_to(groups)

for lat, lon, label in zip(latitudes, longitudes, labels):
    fo.Marker([lat, lon], popup=label).add_to(la_map)
    
la_map.add_child(incidents)
la_map.add_child(groups)
```
